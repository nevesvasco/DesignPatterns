```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, average_precision_score,
                             RocCurveDisplay, PrecisionRecallDisplay)

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Reprodutibilidade
tf.random.set_seed(42)
np.random.seed(42)
```

```python
from google.colab import files
import io

# Inform the user to upload the file
print("Please upload the 'creditcard.csv' file.")
uploaded = files.upload()

df = pd.DataFrame() # Initialize df as an empty DataFrame

# Check if 'creditcard.csv' was uploaded
if 'creditcard.csv' in uploaded:
    df = pd.read_csv(io.BytesIO(uploaded['creditcard.csv']))
    print("File 'creditcard.csv' uploaded and loaded successfully.")
elif uploaded:
    # If a file was uploaded but not 'creditcard.csv', inform the user
    uploaded_filename = list(uploaded.keys())[0]
    print(f"Warning: A file named '{uploaded_filename}' was uploaded instead of 'creditcard.csv'.")
    print("Please ensure you upload 'creditcard.csv' or adjust the filename in the code.")
    # Attempt to read the first uploaded file if it's the only one
    if len(uploaded) == 1:
        print(f"Attempting to read '{uploaded_filename}' anyway.")
        df = pd.read_csv(io.BytesIO(uploaded[uploaded_filename]))
    else:
        print("Multiple files were uploaded and 'creditcard.csv' was not among them. Please adjust the code to specify which file to use.")
else:
    print("No file was uploaded. Please upload 'creditcard.csv' to proceed.")

# Proceed with the rest of the original code only if df was successfully loaded
if not df.empty:
    print(df.shape)
    print(df.isnull().sum())
    print(df["Class"].value_counts())

    # Visualização do desbalanceamento
    sns.countplot(x="Class", data=df)
    plt.title("Distribuição das Classes (0=Legítimo, 1=Fraude)")
    plt.show()

    # Distribuição do Amount por classe
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    df[df["Class"] == 0]["Amount"].hist(ax=axes[0], bins=50, log=True)
    axes[0].set_title("Amount — Legítimo")
    df[df["Class"] == 1]["Amount"].hist(ax=axes[1], bins=50, log=True, color="red")
    axes[1].set_title("Amount — Fraude")
    plt.show()
else:
    print("DataFrame not loaded. Please address the file upload issue and re-run the cell.")
```
